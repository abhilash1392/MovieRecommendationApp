```python
import numpy as np 
import pandas as pd 
import re
```

```python
df_data  = pd.read_csv('u.data',delimiter='\t',header=None,names=['user_id','movie_id','rating','timestamp'])
df_data.head()
```

```python
df_movie_titles = pd.read_csv('Movie_Id_Titles',delimiter=',')
df_movie_titles.columns = ['movie_id','title']
df_movie_titles.head()
```

```python
df_data_new =df_data.merge(df_movie_titles)
df_data_new.head()
```

```python
def removing_year(x):
    x = x[:-7]
    return x
```

```python
df_data_new['movie_title']=df_data_new['title']
df_data_new.head()
```

```python
df_data_new['movie_title']=df_data_new['movie_title'].apply(lambda x:removing_year(x))
df_data_new.head()
```

```python
df_data = df_data_new.drop(['timestamp','title'],axis=1)
df_data
```

```python
df_data.groupby('movie_title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df_data.groupby('movie_title')['rating'].count()
```

```python
data = pd.DataFrame(df_data.groupby('movie_title')['rating'].mean())
data['rating_count'] = df_data.groupby('movie_title')['rating'].count()
```

```python
data
```

```python
moviemat = df_data.pivot_table(index='user_id',columns='movie_title',values='rating',aggfunc=np.mean)
moviemat
```

```python
movie_name_user_ratings = moviemat['Star Wars']
movie_name_user_ratings
```

```python
similar_to_the_movie = moviemat.corrwith(movie_name_user_ratings)
similar_to_the_movie
```

```python
corr_movie = pd.DataFrame(similar_to_the_movie,columns=['correlation'])
corr_movie
```

```python
corr_movie = corr_movie.join(data)
corr_movie
```

```python
recommended_movies = corr_movie[corr_movie['rating_count']>100].sort_values('correlation',ascending=False).head(6)
recommended_movies
```

```python
recommended_movies.reset_index(inplace=True)
recommended_movies
```

```python
recommended_movies.columns = ['Title','Correlation','Rating','No of Ratings']
recommended_movies
```

```python
recommended_movies.drop('Correlation',inplace=True,axis=1)
print(recommended_movies)
```
